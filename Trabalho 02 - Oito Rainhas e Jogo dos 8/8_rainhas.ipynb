# Problema 8 rainhas

O problema consiste em posicionar todas as 8 rainhas em posições do tabuleiro de modo que nenhuma rainha seja capaz de matar outra rainha. Cada rainha pode realizar os seguintes movimentos.

```python
def diagonal_secundaria(tabuleiro, linha, coluna):
    '''
        Captura a diagonal secundaria de uma posição da matriz (linha, coluna)
    
        tabuleiro: matriz 8x8 correspondente a o estado atual do tabuleiro
        linha: linha da matriz do campo analisado
        coluna: coluna da matriz do campo analisado
    
        retorna um array contendo todos os numeros da diagonal secundária
    '''
    diagonal = []
    
    l = linha
    lista = []
    for c in range(coluna, len(tabuleiro)):
        lista.append(tabuleiro[l,c])
        l -= 1
        
    diagonal.extend(lista[::-1])
    
    lista = []
    c = coluna
    for l in range(linha + 1, len(tabuleiro)):
        c -= 1
        lista.append(tabuleiro[l,c])
        
    diagonal.extend(lista) 
    
    return np.array(diagonal)
```

```python
class No:
    '''
        Cria um nó na arvore
        
        Atributos
        tabuleiro: Matriz 8x8 corespondente ao estado atual do tabuleiro com as rainhas
        linha: linha onde deve ser inserida a nova rainha
    '''
    def __init__(self, tabuleiro, linha):
        self.tabuleiro = tabuleiro
        self.linha = linha
        self.filhos = []
        
    def cria_filhos(self):
        '''
            Cria para cada nó os filhos possíveis
        '''
        for coluna in range(len(self.tabuleiro)):
            if self.posicao_valida(coluna):
                # essa posição pode inserir uma rainha
                tab = self.tabuleiro.copy()
                tab[self.linha][coluna] = 1
                
                self.filhos.append(No(tab, self.linha+1))
                
    def posicao_valida(self, coluna):
        '''
        Verifica se uma posição dentro do tabuleiro é valida para uma rainha

        tabuleiro: matriz 8x8 correspondente ao tabuleiro de xadrex
        linha: linha no tabuleiro analisada
        coluna: coluna no tabuleiro analisada
        '''
        # verificar se existe alguma rainha na linha
        if 1 in self.tabuleiro[self.linha, :]:
            return False

        # verificar se existe alguma rainha na coluna
        if 1 in self.tabuleiro[:, coluna]:
            return False

        # verificar se existe alguma rainha na diagonal principal
        diagonal_principal = np.diagonal(self.tabuleiro, coluna - self.linha)
        if 1 in diagonal_principal:
            return False

        # verificar se existe alguma rainha na diagonal secundaria
        diagonal_secundaria_valor = diagonal_secundaria(self.tabuleiro, self.linha, coluna)
        if 1 in diagonal_secundaria_valor:
            return False

        return True    
```

```python
def criar_arvore():
    '''
    Cria o no raiz e uma fila de criação dos filhos de cada nó
    
    Também encontra as soluções enquanto cria os filhos
    
    Retorna a arvore e uma lista de matrizes com as soluções possiveis 
    '''
    
    tabuleiro = np.zeros((8,8))
    raiz = No(tabuleiro, 0)
    fila = [raiz]
    solucoes = []
    
    while fila:
        no = fila.pop(0)
        
        # capturar solução
        if np.sum(no.tabuleiro) == 8:
            solucoes.append(no.tabuleiro)
        
        # criar filhos
        if no.linha < 8:
            no.cria_filhos()
            fila.extend(no.filhos)
            
    return raiz, solucoes
```

```python
arvore, solucoes = criar_arvore()
```

```python
def plotar_tabuleiro(sol):
    n = sol.shape[0]

    tabuleiro = plt.figure(figsize=(5,5))
    ax = tabuleiro.add_subplot(111, aspect='equal')

    for i in range(n):
        for j in range(n):
            cor = 'white' if (i + j) % 2 == 0 else 'black'
            ax.add_patch(plt.Rectangle((i, j), 1, 1, color=cor))

    for i in range(n):
        for j in range(n):
            if sol[i, j] == 1:
                ax.plot(j + 0.5, i + 0.5, 'ro', markersize=15)
    
    plt.show()
```

```python
for solucao in solucoes:
    plotar_tabuleiro(solucao)
```
